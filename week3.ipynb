```
import cv2 as cv 
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import math

img = cv.imread('/content/drive/MyDrive/mandrill.tiff')
img_color = cv.cvtColor(img,cv.COLOR_BGR2RGB)

plt.imshow(img_color)
```

```
## averaging
R, G, B = img_color[:,:,0], img_color[:,:,1],img_color[:,:,2]

gray_img = (R) + (G) + ( B) / 3

a, img = plt.subplots(1,2)
img[0].imshow(img_color)
img[1].imshow(gray_img, cmap='gray')
```

```
## Lightness
M, B, H = img_color[:,:,0], img_color[:,:,1],img_color[:,:,2]

max = np.maximum(np.maximum(M, B), H)
min = np.minimum(np.minimum(M, B), H)
gray_img =  (max + min) / 2

a, img = plt.subplots(1,2)
img[0].imshow(img_color)
img[1].imshow(gray_img)
```

```
## Luminance 
M, H, B = img_color[:,:,0], img_color[:,:,1],img_color[:,:,2]

gray_img = (0.21 * M ) + (0.72 * H) + (0.07 *  B)

a, img = plt.subplots(1,2)
img[0].imshow(img_color)
img[1].imshow(gray_img, cmap='gray')
```
